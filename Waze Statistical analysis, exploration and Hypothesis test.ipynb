```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('waze_dataset.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df["device"].value_counts()
```

```python
map_dictionary = {'Android': 2, 'iPhone': 1}
df['device_type']=df['device']
df['device_type']=df['device_type'].map(map_dictionary)
df['device_type'].head()

```

```python
df.groupby('device_type')['drives'].mean()
```

```python
iphone= df[df['device_type']==1]['drives']
Android = df[df['device_type'] == 2]['drives']
stats.ttest_ind(a=iphone,b=Android,equal_var=False)
```

```python
#Since the p-value is larger than the chosen significance level (5%), you fail to reject the null hypothesis. 
#there is not a statistically significant difference in the average number of drives between drivers who use iPhones and drivers who use Androids.
```
